```python
from keras_segmentation.pretrained import pspnet_50_ADE_20K , pspnet_101_cityscapes, pspnet_101_voc12

# model = pspnet_50_ADE_20K() # load the pretrained model trained on ADE20k dataset

# model = pspnet_101_cityscapes() # load the pretrained model trained on Cityscapes dataset

model = pspnet_101_voc12() # load the pretrained model trained on Pascal VOC 2012 dataset
```

```python

# load any of the 3 pretrained models
import time
from matplotlib import pyplot as plt
import numpy as np
import cv2

st = time.time()
out = model.predict_segmentation(
    inp="./images/Human.jpeg",
    out_fname="out.png"
)
print(time.time() - st)

img = cv2.imread("./images/Human.jpeg")

plt.figure()
plt.imshow(img)

img = cv2.imread("out.png")
plt.figure()
plt.imshow(img)


st = time.time()
out = model.predict_segmentation(
    inp="./images/cat.jpeg",
    out_fname="out.png"
)
print(time.time() - st)

img = cv2.imread("./images/cat.jpeg")

plt.figure()
plt.imshow(img)

img = cv2.imread("out.png")
plt.figure()
plt.imshow(img)




st = time.time()
out = model.predict_segmentation(
    inp="./images/item_1.jpg",
    out_fname="out.png"
)
print(time.time() - st)

img = cv2.imread("./images/item_1.jpg")

plt.figure()
plt.imshow(img)

img = cv2.imread("out.png")
plt.figure()
plt.imshow(img)




st = time.time()
out = model.predict_segmentation(
    inp="./images/cat_2.jpg",
    out_fname="out.png"
)
print(time.time() - st)

img = cv2.imread("./images/cat_2.jpg")

plt.figure()
plt.imshow(img)

img = cv2.imread("out.png")
plt.figure()
plt.imshow(img)

```

```python

# load any of the 3 pretrained models
import time
from matplotlib import pyplot as plt
import numpy as np
import cv2

cap = cv2.VideoCapture(0)
while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()
    
    # save image to fs
    cv2.imwrite('in.png', frame) 

    out = model.predict_segmentation(
        inp="./in.png",
        out_fname="out.png"
    )
    image = cv2.imread('./out.png') 
    # Display the resulting frame
    vis = np.concatenate((image, frame), axis=1)
    cv2.imshow('frame', vis)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything done, release the capture
cap.release()

```
