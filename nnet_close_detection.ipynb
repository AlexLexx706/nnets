# Простая сеть для классификации обьектов:

```python
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import utils
import numpy as np

(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# преобразование размерности изображений
x_train = x_train.reshape(60000, 784).astype(float)
# print(x_train.flags)

# нормалзация данных
x_train /= 255.
# y_train = y_train.astype(float)
# y_train.dtype

# print(y_train)
#преобразуем метки в категории
y_train = utils.to_categorical(y_train, 10)
# print(y_train)


# подготовка тестовых данных
x_test = x_test.reshape(10000, 784).astype(float)
x_test /= 255.
y_test = utils.to_categorical(y_test, 10)

#названия класов
classes = ['футболка', 'брюки', 'свитер', 'платье', 'пальто', 'туфли', 'рубашка', 'кросовки', 'сумка', 'ботинки',]



```

```python
#создаём последовательную модель
model = Sequential()
model.add(Dense(800, input_dim=784, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(10, activation='softmax'))

#компилируем модель
model.compile(loss='categorical_crossentropy', optimizer='SGD', metrics=['accuracy'])
print(model.summary())
```

```python
#обучаем сеть
model.fit(x_train, y_train, batch_size=400, epochs=100, verbose=1, validation_split=0.2)
```

```python
#запускаем сеть на входных данных
prediction = model.predict(x_train)

#выводим один из результатов распознования
print(prediction[0])

#выводим номер класса
print(np.argmax(prediction[0]))

#выводим правильный номер класса
print(np.argmax(y_train[0]))
```

```python
# проверка качества работы на тестовых данных
scores = model.evaluate(x_test, y_test, verbose=1)
print("Доля правильных ответов на тестовом наборе в процентах:%s" % (round(scores[1] * 100, 4)))
```
